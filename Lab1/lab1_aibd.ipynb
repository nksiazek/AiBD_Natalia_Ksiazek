### Zadanie 3.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from dataclasses import make_dataclass
```

```python
y = lambda x: x**2 + 5  #wyrażenie określające funkcję

#funkcja do wykonywania wykresów
def wykres(x, y, title):
    plt.show()
    plt.plot(x, y, label='f(x)')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()

x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)

y1 = y(x1)
y2 = y(x2)
y3 = y(x3)

wykres(x1, y1, 'Przedział [-1, 1]')
wykres(x2, y2, 'Przedział [-6, 6]')
wykres(x3, y3, 'Przedział [0, 5]')
```

### Zadanie 4.

```python
PersonalData = make_dataclass('PersonalData', [('Name', str), ('Surname', str), ('Age', int), ('Sex', str)])

df = pd.DataFrame([PersonalData('Jan', 'Nowak', 29, 'male'),
PersonalData('Anna', 'Kowalska', 56, 'female'),
PersonalData('Walezy', 'Wolny', 19, 'male'),
PersonalData('Stefan', 'Stok', 29, 'male'),
PersonalData('Monika', 'Moskal', 29, 'female')])
```

```python
print(df) #Wyświetlenie całego DataFrame
print(df.info()) #Wyświetlenie informacji o danych
print(df.describe()) #Wyświetlenie opisu danych
print(df.head(3)) #Wyświetlenie 3 pierwszych rekordów
```
